# The PyData stack

We've already seen how the combination of numpy and matplotlib can create a Matlab-like environment. However, these packages are primarily meant to deal with homogeneous data, mainly numerical.

The main changes to the stack that enabled heterogeneous data was the development of the `pandas` package. This package is built on top of `numpy`, and allows a `DataFrame` object to be created along the lines of R's `data.frame`. There was a lot of functionality included in this package in the form of wrappers to make many operations, including plotting, pretty easy.

### Series

A pandas Series object looks like a numpy vector, but it has one very crucial difference: it has an index which need not be the natural numbers. 

```python
rng = np.random.RandomState(30)
x = rng.random_sample(10)
x
```

```python
x_series = pd.Series(x)
x_series
```

The index can be made into any arbitrary scalar quantities

```python
x_series.index
```

```python
x_series.index = pd.Index(['a','b','c','d','e','f','g','h','i','j'])
```

```python
x_series
```

```python
x_series['e':'h']
```

__Note:__ In pandas, the index-based slicing includes both start and end. Recall in numpy it includes the start but not the end

```python
x_series[:'f']
```

```python
x[3:5]
```

```python
x_series[3:5]
```

```python
x_series.iloc[3:5]
```

```python
x_series.describe()
```

```python
x_series.plot(kind='hist');
```

```python
y = rng.randint(0,10, 10)
y
```

```python
y_series = pd.Series(y)
```

```python
pd.concat([x_series,y_series], axis=1)
```

```python
x_series.index = pd.Index(np.arange(0,10))
```

```python
x_series
```

```python
out=pd.concat([x_series, y_series], axis=1)
out.columns = ['x','y']
out
```

```python
x_series.index = pd.Index(['a','b','c','d','e','f','g','h','i','j'])
out = pd.DataFrame({'x': x_series.values, 'y': y_series.values})
```

### Split-apply-combine

![](SplitApplyCombine.png)

```python
iris.groupby('Species')[['Sepal.Length','Sepal.Width']].mean()
```

```python
iris[(iris['Sepal.Length']>5) & (iris['Sepal.Width'] < 2.7)]
```

```python
def myrange(x):
    return(x.max() - x.min())
```

```python
iris.loc[:,:'Petal.Width'].agg(myrange)
```

```python
iris.groupby('Species').agg([ np.mean, np.std,myrange])
```

### Merges and joins

```python
def make_df(cols, ind):
    data = {c: [str(c)+str(i) for i in ind] for c in cols}
    return(pd.DataFrame(data, ind))

make_df('ABC', range(3))
```

```python
x = [1,2,3]
y = [4,5,6]
z = [7,8,9]
np.concatenate([x,y,z])
```

```python
df1 = make_df('ABC',[1,2])
df2 = make_df('BCD',[3,4])
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join='inner', ignore_index=False)
```

```python
A = pd.DataFrame({'employee':['Andrew','John','Joyce','Sue'],
                 'group': ['Accounting','Engineering','Engineering','HR']})
B = pd.DataFrame({'employee': ['Andrew','Joyce','Sue','John'],
                 'hire_date':[2004, 2010, 2007, 2009]})
```

```python
A
```

```python
B
```

```python
pd.merge(A,B)
```

```python
C = pd.DataFrame({'group':['Accounting','Engineering','HR'],
                'supervisor': ['Carl','Tracy','Steve']})
pd.merge(A,C)
```

```python
payroll = pd.DataFrame({'name': ['Andrew','John','Joyce','Sue'],
                       'salary': [45000, 75000, 80000, 50000]})
payroll
```

```python
pd.merge(A, payroll, left_on='employee',right_on = 'name').drop('name', axis=1)
```

#### US States data

```python
population = pd.read_csv('http://ddlteaching.github.io/FreddieMac/data/state-population.csv')
area = pd.read_csv('http://ddlteaching.github.io/FreddieMac/data/state-areas.csv')
abbrevs = pd.read_csv('http://ddlteaching.github.io/FreddieMac/data/state-abbrevs.csv')
```

```python
population.head()
```

```python
area.head()
```

```python
abbrevs.head()
```

```python
df1 = pd.merge(population, abbrevs, left_on = 'state/region', right_on='abbreviation').drop('abbreviation',axis=1)
df1
```

```python
df2 = pd.merge(df1, area)
```

```python
df2
```

```python
df3=df2[df2['ages']=='total']
```

```python
df3.head()
```

```python
df3['density'] = df3['population']/df3['area (sq. mi)']
df3.head()
```

```python
df4 = df3.loc[df3['state']=='New York',['year','density']]
plt.scatter(df4['year'],df4['density'])
```
